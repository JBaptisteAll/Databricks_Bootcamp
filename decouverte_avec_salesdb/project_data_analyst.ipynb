```sql
-- List all tables
SHOW TABLES IN workspace.salesdb
```

```sql
-- Inspect table structure dim_customer
DESCRIBE TABLE workspace.salesdb.dim_customers
```

```sql
-- List unique customer countries
SELECT 
  DISTINCT(country)
FROM workspace.salesdb.dim_customers
```

```sql
-- List product categories, subcategories, and products
SELECT
  category,
  subcategory,
  product_name
FROM workspace.salesdb.dim_products
```

```sql
-- Calculate: Total sales, Total customers, Total products, Total orders
SELECT
  COUNT(c.customer_id) AS Total_customers,
  SUM(sales_amount) AS Total_sales,
  COUNT(product_id) AS Total_products,
  COUNT(DISTINCT(order_number)) AS Total_orders
FROM workspace.salesdb.fact_sales AS s
JOIN workspace.salesdb.dim_customers AS c USING (customer_key)
JOIN workspace.salesdb.dim_products AS p USING (product_key)
```

```sql
-- Sales by product
SELECT
  p.product_name,
  SUM(s.sales_amount)
FROM workspace.salesdb.fact_sales AS s
INNER JOIN workspace.salesdb.dim_products AS p USING (product_key)
GROUP BY p.product_name
```

```sql
-- Sales by category
SELECT
  p.category,
  SUM(s.sales_amount)
FROM workspace.salesdb.fact_sales AS s
INNER JOIN workspace.salesdb.dim_products AS p USING (product_key)
GROUP BY p.category
```

```sql
-- Orders over time
SELECT
  date_trunc('month', order_date) AS order_month,
  COUNT(DISTINCT(order_number)) AS total_nb_order,
  SUM(sales_amount) AS total_sale
FROM workspace.salesdb.fact_sales
GROUP BY date_trunc('month', order_date)
```

```sql
-- Calculate: Total sales, Total customers, Total products, Total orders
SELECT
  *
FROM workspace.salesdb.fact_sales AS s
JOIN workspace.salesdb.dim_customers AS c USING (customer_key)
JOIN workspace.salesdb.dim_products AS p USING (product_key)
```
